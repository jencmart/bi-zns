```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
%matplotlib inline

import warnings
warnings.filterwarnings('ignore') # haha. I figured out solution for all warings.

# moje 

# universum
# univerza předpokladů
vaha = ctrl.Antecedent(np.arange(0, 601, 10), 'Vaha v gramech')
kvalita_zpracovani = ctrl.Antecedent(np.arange(0, 11, 1), 'kvalita zpracovani')
kvalita_fotek = ctrl.Consequent(np.arange(0, 11, 1), 'kvalita fotek')
velikost_ram = ctrl.Consequent(np.arange(0, 2001, 10), 'RAM v MB')
frekvence_cpu = ctrl.Consequent(np.arange(0, 2401, 100), 'CPU v MHz')

# fuzzy fce
# fce prislusnosti 
vaha["mala"] = fuzz.trapmf(vaha.universe,[0,0,300,330])
vaha["stredni"] = fuzz.trapmf(vaha.universe,[250,350,400,500])
vaha["velka"] = fuzz.trapmf(vaha.universe,[350,420,600,600])

kvalita_zpracovani["hrozna"] = fuzz.trapmf(kvalita_zpracovani.universe,[0,0,3,4])
kvalita_zpracovani["prumerna"] = fuzz.trapmf(kvalita_zpracovani.universe,[2,4,5,7])
kvalita_zpracovani["spickova"] = fuzz.trapmf(kvalita_zpracovani.universe,[5,8,10,10])

kvalita_fotek["brambora"] = fuzz.trapmf(kvalita_fotek.universe,[0,0,2,3])
kvalita_fotek["spatna"] = fuzz.trapmf(kvalita_fotek.universe,[2,3,4,5])
kvalita_fotek["prumerna"] = fuzz.trapmf(kvalita_fotek.universe,[4,5,7,8])
kvalita_fotek["perfektni"] = fuzz.trapmf(kvalita_fotek.universe,[7,8,10,10])

velikost_ram["nedostatecna"] = fuzz.trapmf(velikost_ram.universe,[0,0,200,600])
velikost_ram["prumerna"] = fuzz.trapmf(velikost_ram.universe,[200,660,800,1000])
velikost_ram["nadprumerna"] = fuzz.trapmf(velikost_ram.universe,[700,1400,2000,2000])


frekvence_cpu["mala"] = fuzz.trapmf(frekvence_cpu.universe,[0,0,200,500])
frekvence_cpu["stredni"] = fuzz.trapmf(frekvence_cpu.universe,[300,600,800,1000])
frekvence_cpu["velka"] = fuzz.trapmf(frekvence_cpu.universe,[800,1200,2400,2400])

# zobraz
vaha.view()
kvalita_zpracovani.view()
kvalita_fotek.view()
velikost_ram.view()
frekvence_cpu.view()
```
